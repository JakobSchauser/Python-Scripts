```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import seaborn as sns
sns.set()



tohit = 7
dmg = 10.1

grit = 0

n = int(1e5)
ACS = np.array(range(10,26))
dmgs = np.zeros((len(ACS),2))

r = np.random.randint(1,21,n) 

for i,AC in enumerate(ACS):
    dmgs[i,0] = np.sum(r+tohit+grit>=AC)*(dmg+grit)
    dmgs[i,1] = np.sum(r+tohit+grit-5>=AC)*(dmg+grit+10)


plt.figure(figsize=(14,6))

plt.plot(ACS,dmgs[:,1]/dmgs[:,0],'.')

plt.hlines(1,ACS[0],ACS[-1],linestyles="--")
plt.xlabel("AC")
plt.ylabel("Improvement")
plt.show()
```
