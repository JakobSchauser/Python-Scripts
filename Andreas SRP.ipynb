```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def simulate(K,L,Sk,R,N = 0,T = 1000):
    ns = [N]
    ts = [0]
    sinus = np.sin(R*np.pi/75+np.pi/2)
    print(sinus)
    K -= R*0.008*sinus
    t=0
    dt=1
    while t<T:
        Sf = (K-1)*N/L
        t += dt
        N += Sk*dt + Sf*dt
        ns.append(N)
        ts.append(t)
    return (ts,ns,K)

def visualize(sim):
    fig, ax = plt.subplots(figsize = (8,5))
    
    ax.plot(sim[0],sim[1],'r.')
    ax.set_title("Simulation of pile with K="+str(sim[2]))
    ax.set_ylabel("Neutron count [m$^{-3}$]")
    ax.set_xlabel("Time [s]")
    
```

```python
ends= []
Rs = np.array(range(0,50))
for i in Rs:
    sim = simulate(0.9995,0.1,1e5,i)
    visualize(sim)
    ends.append(sim[1][-1])
plt.figure()
plt.plot(Rs,1/np.array(ends),'.')
```

```python
visualize(simulate(1,0.1,1e5,0))
```
