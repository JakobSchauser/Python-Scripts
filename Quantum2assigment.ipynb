```python
import numpy as np
import matplotlib.pyplot as plt

% matplotlib notebook

Ep = lambda d1,d2,g: ((d1+d2)+np.sqrt((d1+d2)**2-4*d1*d2+4*g**2))/2
Em = lambda d1,d2,g: ((d1+d2)-np.sqrt((d1+d2)**2-4*d1*d2+4*g**2))/2

d2s = np.linspace(0,2,200)

plt.figure()
plt.plot(d2s,Ep(1,d2s,0),label="g = 0, Positive")
plt.plot(d2s,Em(1,d2s,0),label="g = 0, Negative")
plt.plot(d2s,Ep(1,d2s,0.1),label="g = 0.1, Positive")
plt.plot(d2s,Em(1,d2s,0.1),label="g = 0.1, Negative")
plt.legend()
```

```python
pertubated = lambda d1,d2,g: g**2/(d2-d1)
plt.figure()
#d2s.remove(0)
y = pertubated(1,d2s,0.1)
y[d2s==0]
plt.plot(d2s,-y+1,label="pertubated, Positive",ls='--')
plt.plot(d2s,y+d2s,label="pertubated, Negative",ls='--')
plt.plot(d2s,Ep(1,d2s,0.1),label="g = 0.1, Positive")
plt.plot(d2s,Em(1,d2s,0.1),label="g = 0.1, Negative")
plt.legend()
```
