```python
import numpy as np

def setup():
    deck = np.zeros((52,2)).astype(int)
    deck[:,0] = np.tile(np.array([i for i in range(13)]),4)
    deck[:,1] = np.array([np.floor(i/13) for i in range(52)])
    np.random.shuffle(deck)

    stack = []
    
    return deck,stack

def nextcard(deck,stack):
    last, deck = deck[-1], deck[:-1]
    stack.append(last)
    return deck,stack
    

def remove_from_stack(lis,i):
    return [lis[x] for x in range(len(lis)) if x != i]

def check_moves(stack,verbose = False):
    for i in range(len(stack)):
        n = len(stack)-i-1
        num, suit = stack[n]
        if(n-1 >= 0):
            cnum, csuit = stack[n-1]
            if verbose:
                print(num,cnum,suit,csuit)
            if(cnum == num or csuit == suit):
                stack[n-1] = stack[n] 
                
                stack = remove_from_stack(stack,n) 
                
                return True,stack
        if(n-3 >= 0):
            cnum, csuit = stack[n-3]
            if verbose:
                print(num,cnum,suit,csuit)
            if(cnum == num or csuit == suit):
                stack[n-3] = stack[n] 
                stack = remove_from_stack(stack,n)
                return True,stack
        
    return False,stack

def simulate(runs,verbose=False):
    ntrue = 0
    for run in range(runs):
        deck, stack = setup()
        for i in range(len(deck)):
            
            deck,stack = nextcard(deck,stack)
            if(verbose):
                print("Draw",stack)
            
            still_moving = True
            while still_moving:
                still_moving,stack = check_moves(stack)
            if(verbose):
                print("Move:",stack)
        if(len(stack) < 2):
            ntrue += 1

    return ntrue/runs*100

# simulate(100000)
```

```python
simulate(1,verbose = True)
```
