```python
import sounddevice as sd
import numpy as np
import time
import matplotlib.pyplot as plt
fs = 48000
duration = 1
sd.default.samplerate = fs

lim = 1e-5


def delay(s,n,fadeouttime = 1):
    fadeouttime*=n
    l = len(s)
    m = np.max(s)
    lm = l*n
    v = np.empty(lm)
    for i in range(n):
        v[i*l:(i+1)*l] = s
    for i in range(len(v)):
        v[i] /= (1+(i*fadeouttime)/lm)
    return v

def reverse(s):
    return np.flip(s)

def wavedelay(s,n):
    return delay(np.concatenate([s,reverse(s)]),n)

def reversedelay(s,n):
    return delay(reverse(s),n)

########Record############
timebefore = 4

print(f"recording in {timebefore} seconds!")
time.sleep(timebefore)

print("Recording!")
rec = sd.rec(int(duration * fs),channels=2)
sd.wait()
rec = np.max(rec,axis=1)
clean = rec[abs(rec)>lim]



########Do effects######

# re = wavedelay(clean,3)
re = delay(clean,5)
print("Done!")

plt.plot(rec)
plt.show()
plt.plot(clean)
plt.show()
plt.plot(re)
plt.show()
```

```python
sd.wait()
print("Playing!")
sd.play(re)
```
