```python
import requests
from bs4 import BeautifulSoup
```

```python
getpage = lambda institute, page: f"https://karakterstatistik.stads.ku.dk/#searchText=&term=&block=&institute={institute}&faculty=1868&searchingCourses=true&page={page}"

studies =  [{"name":"bio", "code": 65718, "n_pages":3},
            {"name":"data", "code": 1932, "n_pages":3},
            {"name":"idræt", "code": 1950, "n_pages":3},
            {"name":"matematik", "code": 1869, "n_pages":3},
            {"name":"kemi", "code": 1568, "n_pages":3},
            {"name":"fysik", "code": 65707, "n_pages":3},
          ]
```

```python

stud = studies[0]

for page in range(1,1+stud["n_pages"]):
    response = requests.get(getpage(stud["code"],page), timeout=(3.05, 27))
    soup = BeautifulSoup(response.content, "html.parser")
#     container = soup.find("div", id="searchResults")
    

#     print(soup.find_all('tr'))
    print(soup)
    
    break
    
```
