```python
from xml.dom import minidom
import webbrowser
from time import sleep
```

```python
elements = minidom.parse(open("drum_sitemap.xml")).getElementsByTagName("loc")

locs = []
for e in elements:
    locs.append(e.childNodes[0].data+"?orderby=price")

for l in locs:
    page = webbrowser.open(l)
    sleep(0.3)
#     break
```

```python
import selenium
from selenium import webdriver
import time
from tqdm.notebook import tqdm


driver = webdriver.Chrome()

for i in tqdm(range(10000)):
    driver.get("https://interaktiv.mx.dk/simplevotes/get/11592")
    yes = driver.find_element_by_id("vote_yes")

    try: 
        yes.click()
    except:
        print("Too fast at",i)
    driver.delete_all_cookies()
    time.sleep(0.2)

    
```
