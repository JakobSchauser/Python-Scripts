```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib

def step(game):
    step = np.empty(game.shape).astype(int)
    for x in range(len(step)):
        for y in range(len(step)):
            nbs = np.sum(game[x-1:x+2,y-1:y+2])-game[x,y]
            step[x,y] = nbs
            if(nbs < 2 or nbs > 3):
                step[x,y] = 0
            elif(nbs == 3):
                step[x,y] = 1
            else:
                step[x,y] = game[x,y]
    return step

def new_game(size):
    return np.random.choice([0,1],(size,size))

def run(steps,size):
    fig,ax = plt.subplots()
    game = new_game(size)
    canvas = ax.imshow(game)
    for _ in range(steps):
        game = step(game)
        canvas.set_data(game)
        plt.pause(0.05)

run(10,10)
```

```python
run(10,10)
```
